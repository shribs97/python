```python
def rail_fence_encrypt(text, key):
    rail = [['\n' for i in range(len(text))]
                for j in range(key)]

    dir_down = False
    row, col = 0, 0

    for i in range(len(text)):
        if row == 0 or row == key - 1:
            dir_down = not dir_down

        rail[row][col] = text[i]
        col += 1

        if dir_down:
            row += 1
        else:
            row -= 1

    result = []
    for i in range(key):
        for j in range(len(text)):
            if rail[i][j] != '\n':
                result.append(rail[i][j])
    return ''.join(result)

def rail_fence_decrypt(cipher, key):
    rail = [['\n' for i in range(len(cipher))]
                for j in range(key)]

    dir_down = None
    row, col = 0, 0

    for i in range(len(cipher)):
        if row == 0:
            dir_down = True
        if row == key - 1:
            dir_down = False

        rail[row][col] = '*'
        col += 1

        if dir_down:
            row += 1
        else:
            row -= 1

    index = 0
    for i in range(key):
        for j in range(len(cipher)):
            if ((rail[i][j] == '*') and
               (index < len(cipher))):
                rail[i][j] = cipher[index]
                index += 1

    result = []
    row, col = 0, 0
    for i in range(len(cipher)):

        if row == 0:
            dir_down = True
        if row == key-1:
            dir_down = False

        if (rail[row][col] != '*'):
            result.append(rail[row][col])
            col += 1

        if dir_down:
            row += 1
        else:
            row -= 1
    return ''.join(result)



```

```python
from Crypto.Cipher import DES
from Crypto.Random import get_random_bytes

def encrypt_message(message, key):
    cipher = DES.new(key, DES.MODE_ECB)
    # Make sure message length is a multiple of 8 (DES block size)
    while len(message) % 8 != 0:
        message += b' '
    ciphertext = cipher.encrypt(message)
    return ciphertext

def decrypt_message(ciphertext, key):
    cipher = DES.new(key, DES.MODE_ECB)
    plaintext = cipher.decrypt(ciphertext)
    return plaintext.rstrip(b' ')

# Example usage:
message = b'Hello, DES!'
key = get_random_bytes(8)  # 8 bytes key for DES
print("Original message:", message)

# Encrypt the message
encrypted_message = encrypt_message(message, key)
print("Encrypted message:", encrypted_message)

# Decrypt the message
decrypted_message = decrypt_message(encrypted_message, key)
print("Decrypted message:", decrypted_message.decode())
```

```python
import hashlib

def calculate_sha1(input_string):
    sha1_hash = hashlib.sha1(input_string.encode()).hexdigest()
    return sha1_hash

# Example usage:
input_string = "Hello, SHA-1!"
sha1_hash = calculate_sha1(input_string)
print("SHA-1 hash of", input_string, ":", sha1_hash)
```

```python
def prepare_input(text):
    # Convert to uppercase and remove non-alphabetic characters
    processed_text = ""
    for char in text.upper():
        if char.isalpha():
            processed_text += char
    return processed_text

def generate_table(key):
    key = prepare_input(key)
    alphabet = "ABCDEFGHIKLMNOPQRSTUVWXYZ"
    table = ""
    key_done = []
    for char in key:
        if char not in key_done:
            table += char
            key_done.append(char)
    for char in alphabet:
        if char not in key_done:
            table += char
            key_done.append(char)
    return table

def generate_pairs(text):
    processed_text = prepare_input(text)
    # If the length of the text is odd, add an X to make it even
    if len(processed_text) % 2 != 0:
        processed_text += "X"
    pairs = []
    for i in range(0, len(processed_text), 2):
        if processed_text[i] == processed_text[i+1]:
            pairs.append(processed_text[i] + "X")
            pairs.append("X" + processed_text[i+1])
        else:
            pairs.append(processed_text[i] + processed_text[i+1])
    return pairs

def find_char(char, table):
    for i in range(5):
        for j in range(5):
            if table[i][j] == char:
                return i, j

def playfair_cipher_encrypt(plain_text, key):
    table = generate_table(key)
    pairs = generate_pairs(plain_text)
    cipher_text = ""
    for pair in pairs:
        if pair[0] == pair[1]:
            cipher_text += pair[0] + pair[1]
            continue
        row1, col1 = find_char(pair[0], table)
        row2, col2 = find_char(pair[1], table)
        if row1 == row2:
            cipher_text += table[row1][(col1 + 1) % 5] + table[row2][(col2 + 1) % 5]
        elif col1 == col2:
            cipher_text += table[(row1 + 1) % 5][col1] + table[(row2 + 1) % 5][col2]
        else:
            cipher_text += table[row1][col2] + table[row2][col1]
    return cipher_text

def playfair_cipher_decrypt(encrypted_text, key):
    table = generate_table(key)
    pairs = generate_pairs(encrypted_text)
    plain_text = ""
    for pair in pairs:
        if pair[0] == pair[1]:
            plain_text += pair[0] + pair[1]
            continue
        row1, col1 = find_char(pair[0], table)
        row2, col2 = find_char(pair[1], table)
        if row1 == row2:
            plain_text += table[row1][(col1 - 1) % 5] + table[row2][(col2 - 1) % 5]
        elif col1 == col2:
            plain_text += table[(row1 - 1) % 5][col1] + table[(row2 - 1) % 5][col2]
        else:
            plain_text += table[row1][col2] + table[row2][col1]
    return plain_text

# Example usage:
plain_text = "HelloPlayfairCipher"
key = "KEYWORD"
encrypted_text = playfair_cipher_encrypt(plain_text, key)
print("Encrypted text:", encrypted_text)
decrypted_text = playfair_cipher_decrypt(encrypted_text, key)
print("Decrypted text:", decrypted_text)
```

```python
import numpy as np

def prepare_input(text, key_length):
    processed_text = ""
    for char in text:
        if char.isalpha():
            processed_text += char.upper()
    while len(processed_text) % key_length != 0:
        processed_text += "X"
    return processed_text

def generate_key_matrix(key):
    key = key.upper().replace(" ", "")
    key_length = len(key)
    matrix_size = int(key_length ** 0.5)
    if matrix_size * matrix_size != key_length:
        raise ValueError("Key length must be a perfect square")
    key_matrix = []
    for char in key:
        key_matrix.append(ord(char) - 65)
    return np.array(key_matrix).reshape(matrix_size, matrix_size)

def generate_text_matrix(text, key_length):
    text_matrix = []
    for char in text:
        text_matrix.append(ord(char) - 65)
    return np.array(text_matrix).reshape(-1, key_length)

def hill_cipher_encrypt(plain_text, key):
    key_matrix = generate_key_matrix(key)
    text_length = len(plain_text)
    key_length = len(key)
    plain_text = prepare_input(plain_text, key_length)
    text_matrix = generate_text_matrix(plain_text, key_length)
    cipher_text = ""
    for i in range(text_length // key_length):
        chunk = text_matrix[i].reshape(-1, len(text_matrix[i]))
        chunk = np.dot(chunk, key_matrix) % 26
        for row in chunk:
            for num in row:
                cipher_text += chr(num + 65)
    return cipher_text

def hill_cipher_decrypt(cipher_text, key):
    key_matrix = generate_key_matrix(key)
    key_matrix = np.linalg.inv(key_matrix)
    key_matrix = (key_matrix * np.linalg.det(generate_key_matrix(key))).astype(int)
    key_matrix = key_matrix % 26
    text_length = len(cipher_text)
    key_length = len(key)
    cipher_text = prepare_input(cipher_text, key_length)
    text_matrix = generate_text_matrix(cipher_text, key_length)
    plain_text = ""
    for i in range(text_length // key_length):
        chunk = text_matrix[i].reshape(-1, len(text_matrix[i]))
        chunk = np.dot(chunk, key_matrix) % 26
        for row in chunk:
            for num in row:
                plain_text += chr(num + 65)
    return plain_text

# Example usage:
plain_text = "Hello, Hill Cipher!"
key = "GYBNQKURP"
encrypted_text = hill_cipher_encrypt(plain_text, key)
print("Encrypted text:", encrypted_text)
decrypted_text = hill_cipher_decrypt(encrypted_text, key)
print("Decrypted text:", decrypted_text)
```

```python
def prepare_input(text):
    processed_text = ""
    for char in text:
        if char.isalpha():
            processed_text += char.upper()
    return processed_text

def vigenere_cipher_encrypt(plain_text, key):
    plain_text = prepare_input(plain_text)
    key = prepare_input(key)
    encrypted_text = ""
    key_index = 0
    for char in plain_text:
        if char.isalpha():
            shift = ord(key[key_index]) - 65
            encrypted_char = chr(((ord(char) - 65 + shift) % 26) + 65)
            encrypted_text += encrypted_char
            key_index = (key_index + 1) % len(key)
        else:
            encrypted_text += char
    return encrypted_text

def vigenere_cipher_decrypt(encrypted_text, key):
    encrypted_text = prepare_input(encrypted_text)
    key = prepare_input(key)
    decrypted_text = ""
    key_index = 0
    for char in encrypted_text:
        if char.isalpha():
            shift = ord(key[key_index]) - 65
            decrypted_char = chr(((ord(char) - 65 - shift) % 26) + 65)
            decrypted_text += decrypted_char
            key_index = (key_index + 1) % len(key)
        else:
            decrypted_text += char
    return decrypted_text

# Example usage:
plain_text = "Hello, Vigenere Cipher!"
key = "KEY"
encrypted_text = vigenere_cipher_encrypt(plain_text, key)
print("Encrypted text:", encrypted_text)
decrypted_text = vigenere_cipher_decrypt(encrypted_text, key)
print("Decrypted text:", decrypted_text)
```
